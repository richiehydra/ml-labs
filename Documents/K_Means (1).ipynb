```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from copy import deepcopy
k=3
import random as rd
```

```python
df = pd.read_csv('kmeans.csv')
print(df)
```

```python
plt.scatter(df["X1"],df["X2"],c='black')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x1 = df['X1'].values
print('x1 - values',x1)
x2 = df['X2'].values
print('x2 - values',x2)
```

```python
X = np.array(list(zip(x1,x2)))
#X=np.array((x1,x2))
print(X)
```

```python
Centroid_y = [6.2, 6.6 , 6.5]
Centroid_x = [3.2, 3.7 ,3.0]
Centroid = np.array(list(zip(Centroid_x, Centroid_y)), dtype=np.float32)
print("Initial Centroids")
print(Centroid.shape)
Centroid
```

```python
Centroid_old = np.zeros(Centroid.shape)
print(Centroid_old)
```

```python
clusters = np.zeros(len(X))
print(clusters)
```

```python
def euclidean_distance(a,b, ax=1):
    return np.linalg.norm(a-b, axis=ax)
```

```python
error = euclidean_distance(Centroid, Centroid_old,None)
print(error)
```

```python
iterr = 0
```

```python
while error != 0:
        # Assigning each value to its closest cluster
        iterr = iterr + 1
        for i in range(len(X)):
            print("Data Points")
            print(X[i])
            distances = euclidean_distance(X[i], Centroid)
            print("Distances")
            print(distances)
            cluster = np.argmin(distances)
            clusters[i] = cluster
        Centroid_old = deepcopy(Centroid)
        print("Old Centroid")
        print(Centroid_old)
        
            
        
        # Finding the new centroids by taking the Mean
        for p in range(k):
            points = [X[j] for j in range(len(X)) if clusters[j] == p]
            Centroid[p] = np.mean(points, axis=0)
        print(" New Centroids after ", iterr," Iteration \n", Centroid)
        error = euclidean_distance(Centroid, Centroid_old, None)
        print("Error ... ",error)
        print("Data points belong to which cluster")
        print(clusters)
        print("********************************************************")
```
